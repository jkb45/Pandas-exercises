### Step 7. Create a dataFrame called vol, with the volume values.

```python
vol=df[['volume_x','volume_y', 'volume']].copy()
vol.head(20)
```

### Step 8. Aggregate the data of volume to weekly.
Hint: Be careful to not sum data from the same week of 2015 and other years.

```python
#logic = {'<Open>'  : 'first',
    #     '<High>'  : 'max',
  #       '<Low>'   : 'min',
   #      '<Close>' : 'last',
   #      '<Volume>': 'sum'}

logic = {'volume': 'sum',
        'volume_x': 'sum', 
        'volume_y':'sum'}

vol.index = pd.to_datetime(vol.index)

vol_agg = vol.resample('W').apply(logic)
#vol.index = vol.index - pd.tseries.frequencies.to_offset("6D")
vol_agg
```
